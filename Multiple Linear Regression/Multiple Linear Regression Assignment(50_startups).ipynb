- Multicollinearity is evident in the data due to a moderate correlation
between research and development expenditure and marketing expenditure.
To assess the extent of multicollinearity,we employ the variance inflation
factor(VIF) as a diagnostic measure.


```python
# p values for all features by fitting model
model.pvalues
```

```python
# checking the effect of R&D on model and comparing its p values
model_rd = smf.ols('Profit~RD_Spend+Administration', data = df).fit()
model_rd.pvalues
```

```python
#checking the effect of administration on model and comparing checking its pvalues
model_ms = smf.ols('Profit~Administration+Marketing_Spend',data = df).fit()
model_ms.pvalues
```

```python
#including vif model from stamodel
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#as we only want to work with feature data we drop the target data
data = df.drop('Profit',axis = 1)
```

```python
#creating the dataframe to store VIF score for various feature variable
vif = pd.DataFrame()
vif['features'] = data.columns
```

```python
#finding VIF score using list comprehension & storing it into dataframe
vif['VIF'] = [variance_inflation_factor(data.values, i) for i in range(len(data.columns))]
vif
```

```python
#training the model using R&D spend features with rest other feature
model_rd = smf.ols('Profit~RD_Spend+Administration',data = df).fit()
model_rd.rsquared,model_rd.aic
```

```python
#training the model using marketing spend features with rest other feature
model_ms = smf.ols('Profit~Administration+Marketing_Spend',data = df).fit()
model_ms.rsquared,model_ms.aic
```

```python
#from the above comparison we can confirm by comparing R squared and aic score that out of two correlated features R&D spend perfroms better, hence we eleminate Marketing spend feature for our final model
scaled_df = pd.DataFrame(sc.fit_transform(df),columns=df.columns)
sns.kdeplot(data=scaled_df)
```

```python
final_model = smf.ols('Profit~RD_Spend+Administration', data = scaled_df).fit()
final_model.summary()
```
